### Interview Questions


### 1.What is Normalization & Standardization and how is it helpful?
Ans: Normalization: Normalization is the process of adjusting values measured on different scales to a common scale, typically between 0 and 1 (or sometimes -1 and 1).Standardization transforms data to have a mean of 0 and a standard deviation of 1, which is more robust to outliers and is ideal for algorithms assuming a normal distribution. 
How is it helpful: Normalization is useful when you need all the data to be within a specific range and donâ€™t assume any particular data distribution. It's typically used with algorithms sensitive to the magnitude of features, such as neural networks and k-nearest neighbors.

### 2.What techniques can be used to address multicollinearity in multiple linear regression?
Ans:To address multicollinearity in multiple linear regression, you can use several techniques. One of the simplest methods is to remove one of the highly correlated predictors after identifying them using a correlation matrix or the Variance Inflation Factor (VIF). Alternatively, you can combine correlated variables into a single new variable, for example, by creating a composite index or using a dimensionality reduction technique like Principal Component Analysis (PCA). Another effective approach is to use regularization techniques such as Ridge or Lasso regression, which penalize the model's coefficients and are designed to handle correlated features. Finally, increasing the sample size can sometimes help by providing more data to better distinguish between the effects of correlated variables.
