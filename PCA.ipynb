### Task 2: Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
```

```python
df_new = pd.DataFrame(SS.fit_transform(df))
df_new.columns = list(df)
df = df_new
df
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pc = pca.fit_transform(df)
pc = pd.DataFrame(pc)
pc.head()
```

```python
 #2.Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.
import numpy as np
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),pca.explained_variance_ratio_, marker='o',linestyle='--')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()
```

### Task 3: Clustering with Original Data:

```python
X = df.iloc[:,1:]
```

```python
from sklearn.cluster import KMeans
K_cluster = KMeans(n_clusters=3)
```

```python
df["K_3"] = K_cluster.fit_predict(X)
```

```python
plt.scatter(df['Alcohol'], df['Proline'], c=df["K_3"], cmap='rainbow')
plt.xlabel('Alcohol')
plt.ylabel('Proline')
plt.show()
```

```python
import numpy as np
from sklearn.metrics import silhouette_score
score=silhouette_score(X,df["K_3"])
print("K_3 score:", np.round(score,3))
```

```python
scores = []
```

```python
for i in range(2,11,1):
    from sklearn.cluster import KMeans
    K_cluster = KMeans(n_clusters=i)
    df["K_3"] = K_cluster.fit_predict(X)
    from sklearn.metrics import silhouette_score
    scores.append(silhouette_score(X,df["K_3"]))
```

```python
np.round(scores,3)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(2,11),scores)    
plt.plot(range(2,11),scores,color='red')
plt.show()
```

2. Similarities and Differences in Clustering Results
   
*Similarities:

3.Impact of Dimensionality Reduction on Clustering Performance

*PCA reduces noise and removes redundant correlations between features,which can help clustering algorithms like K-Means find more stable groups.
*Scores have been significantly reduced after dimensionality reduction on clustering performance.

4. Trade-offs Between Using PCA vs Clustering on Original Dataset

Clustering on Original Dataset:
1.Uses full variance, including noise
2.More sensitive to noisy or irrelevant features
3.Performance & Speed are Slower on high-dimensional data   
4.Cluster Quality may be higher if all features are informative

Clustering after PCA:
1 Keeps only top-variance components (possible info loss) 
2.Reduces noise and redundancy
3.Performance & Speed are Faster due to reduced dimensions
4.Cluster Quality often slightly lower but more stable 

### Task 6: Conclusion and Insights

1.Key Findings and Insights:
  1.Standardization ensured all features contributed equally to the PCA and clustering processes.
  2.PCA successfully reduced the datasetâ€™s dimensionality while retaining most of the variance (~95%), making the data more compact and less noisy.

2.Practical Implications of Using PCA and Clustering:
  1.PCA helps simplify high-dimensional datasets, improves computational efficiency, and reduces the risk of overfitting by removing 
  redundant or noisy features.
  2.Clustering uncovers underlying group structures in data, which can be valuable for segmentation, anomaly detection, and pattern discovery.
  3.Combining PCA and clustering is especially beneficial for large, high-dimensional datasets where noise and feature correlation could 
  hinder cluster stability. 

3.Recommendations:
  1.Use PCA before clustering when working with high-dimensional data, particularly when many features are correlated or contain noise.
  2.Perform clustering directly on the original dataset if the number of features is manageable and all features are expected to be informative.
  3.Always compare clustering results before and after PCA to ensure that the dimensionality reduction has not removed critical separation information.
  4.For practical projects, balance interpretability (original features) against efficiency and stability (PCA features).
