```python
X = df.iloc[:,1:12]
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

### Implementing Clustering Algorithms , Visualization and Evaluation and Performance Metrics:

### K-Means Clustering


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    labels = kmeans.fit_predict(SS_X)
    score = silhouette_score(SS_X, labels)
    print(f"k={i}, Silhouette Score={score:.4f}")
```

```python
from sklearn.cluster import KMeans
K_cluster = KMeans(n_clusters=3)
df["K_3"] = K_cluster.fit_predict(SS_X)
```

```python
plt.scatter(SS_X.iloc[:,0], SS_X.iloc[:,5], c=df["K_3"], cmap='rainbow')  
plt.show()
```

```python
score=silhouette_score(SS_X,df["K_3"])
print("score:", np.round(score,3))
```

### Aglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
linkage_methods = ["ward", "complete", "average", "single"]
for linkage in linkage_methods:
    for n in range(2, 11):
        agg = AgglomerativeClustering(n_clusters=n, linkage=linkage)
        labels = agg.fit_predict(SS_X)
        score = silhouette_score(SS_X, labels)
        print(f"Linkage={linkage}, Clusters={n}, Silhouette Score={score:.4f}")
```
