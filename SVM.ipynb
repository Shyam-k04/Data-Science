### Task 2: Data Preprocessing

```python
df_num1 = df.iloc[:,0:1]
df_num2 = df.iloc[:,24:]
df_num = pd.concat([df_num1,df_num2],axis=1)
df_num
```

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_num = SS.fit_transform(df_num)
SS_num = pd.DataFrame(SS_num)
SS_num.columns = list(df_num)
SS_num
```

```python
df_cat = df.select_dtypes(include=['object', 'category'])
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
LE_cat = df_cat.apply(LE.fit_transform)
LE_cat = pd.DataFrame(LE_cat)
LE_cat.columns = list(df_cat)
LE_cat
```

```python
df_new = pd.concat([SS_num,LE_cat],axis=1)
df_new
```

```python
X = df_new.drop(["class"],axis=1)
Y = df_new["class"]
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pc_X = pca.fit_transform(X)
pc_X = pd.DataFrame(pc_X)
pc_X
```

```python
X_pc = pc_X.iloc[:,0:5]
Y = df_new["class"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_pc,Y,test_size=0.2)
```

### 2. Analysis of SVM Strengths & Weaknesses on Mushroom Dataset

#Strengths:
1.High Accuracy: The Mushroom dataset is categorical and fairly “clean,” so SVMs (especially with the RBF kernel) achieve near-perfect classification.
2.Handles High Dimensionality Well: After encoding, the dataset has many features. SVMs perform well in high-dimensional spaces, especially with kernels like RBF.
3.Robustness to Overfitting: The margin maximization principle helps generalize better, even when classes overlap.

#Weaknesses:
1.Computational Cost: Training SVMs can be slow for very large datasets (many samples). The Mushroom dataset ismanageable, but scaling to millions of records would be problematic.
2.PCA Visualization Limitations: Decision boundaries in 2D PCA space may look imperfect compared to real separation in higher dimensions. Visualizations can sometimes be misleading.
3.Feature Scaling Sensitivity: Though our dataset is categorical (after encoding), in mixed datasets, SVMs can be sensitive to feature scaling, requiring careful preprocessing.

### 3. Practical Implications of SVM in Real-World Classification

1.Medical Diagnosis: Classifying patients as “disease” vs “healthy” (e.g., cancer detection, based on gene expression data).
2.Text & Document Classification: SVMs are powerful in spam detection, sentiment analysis, and topic categorization due to their strength in high-dimensional spaces (like bag-of-words features).
3.Image Recognition:Face detection and handwriting recognition (MNIST dataset) often use SVMs with kernels for feature-rich data.
